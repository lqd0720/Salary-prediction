```python
import pandas as pd
# Define neighborhood to city mappings
CITY_MAPPINGS = {
    # New York boroughs/neighborhoods
    'manhattan': 'New York',
    'brooklyn': 'New York',
    'queens': 'New York',
    'bronx': 'New York',
    
    # Los Angeles area
    'west hollywood': 'Los Angeles',
    'marina del rey': 'Los Angeles',
    'culver city': 'Los Angeles',
    'santa monica': 'Los Angeles',
    'beverly hills': 'Los Angeles',
    'glendale': 'Los Angeles',
    'pasadena': 'Los Angeles',
    
    # San Francisco Bay Area (keeping these distinct)
    'san francisco': 'San Francisco',
    'oakland': 'Oakland',
    'san jose': 'San Jose',
    'palo alto': 'Palo Alto',
    'sunnyvale': 'Sunnyvale',
    
    # Chicago
    'chicago': 'Chicago',
    
    # Other major cities
    'boston': 'Boston',
    'seattle': 'Seattle',
    'austin': 'Austin',
    'dallas': 'Dallas',
    'houston': 'Houston',
    'atlanta': 'Atlanta',
    'miami': 'Miami',
    'philadelphia': 'Philadelphia'
}

def transform_city_name(city):
    """Convert neighborhood names to primary city names"""
    city_lower = city.lower().strip()
    for neighborhood, primary_city in CITY_MAPPINGS.items():
        if neighborhood in city_lower:
            return primary_city
    return city  # Return original if no mapping found

# Load your dataset
df = pd.read_csv('final_data.csv')

# Apply the transformation
df['cities'] = df['cities'].apply(transform_city_name)

# Save the transformed data
df.to_csv('final_data0.csv', index=False)
print(f"Total rows processed: {len(df)}")
```
